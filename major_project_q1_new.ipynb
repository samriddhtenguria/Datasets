```python
# divide data into i/p andd o/p

x = df.iloc[:,2]
x
```

```python
y = df.iloc[:,13]
y
```

```python
# train_test_split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
# Apply TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer # sklearn wala package hai and TfidfVectorizer library hai
vect = TfidfVectorizer() # library ko as a method le rahe hain kyunki isko ek variable mein le rahe hain naa.

x_train_v = vect.fit_transform(x_train) # for training data, it is vect.fit_transform
x_test_v = vect.transform(x_test) # for testing data, it is vect.transform
```

```python
# Apply SVC - SUPPORT VECTOR CLASSIFIER

from sklearn.svm import SVC
model = SVC()
```

```python
# fitting the model
model.fit(x_train_v,y_train) 
```

```python
# Predictor variable

y_pred = model.predict(x_test_v)
y_pred # predicted values
```

```python
y_test  # actual values
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
# Pipelining
from sklearn.pipeline import make_pipeline
text_model = make_pipeline(TfidfVectorizer(),SVC())
text_model.fit(x_train,y_train)
```

```python
# predictor variable
y_pred1 = text_model.predict(x_test)
y_pred1
```

```python
y_test
```

```python
accuracy_score(y_pred1,y_test)*100
```
